```python
a = ['남원읍', '대륜동', '대정읍', '대천동', '동홍동', '서홍동', '성산읍', '송산동', '안덕면', '영천동', '예래동', '정방동', '중문동', '중앙동', '천지동', '표선면', '효돈동']
dfs = list()
for loc in a:
    df = pd.read_csv(f'{loc}.csv')
    df['동'] = '서귀포시' + ' ' + loc
    df.columns = ['날짜', '시간', '습도', '동']
    df = df[df['시간'].isin(['600','700','800','900','1000','1100','1800','1900'])]
    dfs.append(df)

weather = pd.concat(dfs, axis=0)
```

```python
df = pd.pivot_table(weather, index=['동', '날짜'], columns='시간', values='습도').reset_index()
```

```python
날짜=[]
for j in range(17):
    for i in range(30):
        if i<9:
            temp= '2019-09-0' + str(i+1)
        else:
            temp= '2019-09-' +str(i+1)
        
        날짜.append(temp)
    
df['날짜']= 날짜
```

```python
df.to_csv('서귀포시 습도.csv')
```

```python
all_name=['건입동', '구좌읍', '노형동', '도두동', '봉개동', '삼도1동', '삼도2동', '삼양동', '아라동', '애월읍', '연동', '오라동', '용담1동', '용담2동', '우도면', '이도1동', '이도2동', '이호동', '일도1동', '일도2동', '조천읍', '추자면', '한경면', '한림읍', '화북동']
dfs = list()
for loc in all_name:
    df = pd.read_csv(f'{loc}.csv')
    df['동'] = '제주시' + ' ' + loc
    df.columns = ['날짜', '시간', '습도', '동']
    df = df[df['시간'].isin(['600','700','800','900','1000','1100','1800','1900'])]
    dfs.append(df)

weather1 = pd.concat(dfs, axis=0)
```
